```python
import pandas as pd

url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
df = pd.read_csv(url, sep=';').fillna(0)
df
```

```python
#calculando a média de quartos por imoveis
df_media = df.groupby('Tipo')['Quartos'].mean().to_frame().rename(columns={'Quartos': 'Média_Quarto'})
df_media
```

```python
#Conferir quantos bairros únicos existem na nossa base de dados
Df_bairros_unicos = df['Bairro'].nunique()
Df_bairros_unicos
```

```python
# Analisar quais bairros possuem a média de valor de aluguel mais elevadas;
df_bairros_media = df.groupby('Bairro')['Valor'].mean().to_frame().sort_values('Valor',  ascending=False)
df_bairros_media
```

```python
# Criar um gráfico de barras horizontais que apresente os 5 bairros com as médias de valores de aluguel mais elevadas.
df_bairros_media = df.groupby('Bairro')['Valor'].mean().to_frame().sort_values('Valor',  ascending=False).head(5)
df_bairros_media.plot(kind='barh', figsize=(14,10), color='blue'); 
```
